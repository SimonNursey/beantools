```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
```

```python
#log = pd.read_csv("/Users/simon_nursey/go/src/beanex/brew/orderscan/log/bin_03Feb19.csv")
log = pd.read_csv("/Users/simon_nursey/go/src/beanex/bots/kevin/log/eth05Feb19.csv")
log['Position']=log['Position'].fillna(0)
pp = PdfPages('charts.pdf')
for j in range(len(log)//1000+1):
    st=j*1000
    en=min((j+1)*1000,len(log))
    plt.figure(j,figsize=(20,10))
    plt.plot(log['Bid'][st:en],label='Bid',color='Green')
    plt.plot(log['Ask'][st:en],label='Ask',color='Red')
    plt.plot(log['CloseBid'][st:en],label='StaticBid',color='DarkGreen',linewidth=10)
    plt.plot(log['CloseAsk'][st:en],label='StaticAsk',color='DarkRed',linewidth=10)
    plt.ylabel("Price")
    plt.xlabel("Iteration")
    plt.twinx()
    plt.plot(log['PnlEst'][st:en],label='Pnl',color='Blue',linestyle='dotted')
    plt.ylabel("Pnl")
    plt.figlegend(loc='center right')
    posst=st
    for i in range(st+1,en):
        if log['Position'][i]!=0.0 and log['Position'][i-1]==0.0:
            posst=i
        elif log['Position'][i-1]>0.0 and (log['Position'][i]==0.0 or i==en-1):
            plt.axvspan(posst,i,facecolor='b',alpha=0.25)
        elif log['Position'][i-1]<0.0 and (log['Position'][i]==0.0 or i==en-1):
            plt.axvspan(posst,i,facecolor='r',alpha=0.25)
    pp.savefig()
pp.close()
```
